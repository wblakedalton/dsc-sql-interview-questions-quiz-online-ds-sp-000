## Question 2

_______________________________________________________________________________________________________________________________
- Primary keys are unique identifiers for a table and are considered unqiue values for a column entry, while foreign keys are primary keys from another table, allowing for tables to be joined.




## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
- One-to-One relationships occur when a single record in the first table is related to a single record in the second table and vice versa.
- One-to-Many relationships occur when a single record in the first table is related to multiple records in the second table, while a single record in the second table is related to a single record in the first table.
- Many-to-Many relationships occur when a single record in the first table is related to multiple records in the second table and a single record in the second table is related to multiple records in the first table.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
- Inner joins select records from the first table and second table that have matching values.
- Left joins select all records from the first table and the matching values from the second table.
- Right joins select all records from the second table and the matching values from the first table.
- Full joins select all records from the first table and second table when there is a match in either the first or second table.
